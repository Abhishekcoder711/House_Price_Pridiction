```python
import pandas as pd
import numpy as np

from google.colab import files
print(type(files))
files.upload()

df = pd.read_csv("house_prices.csv")
print("house_prices: \n")
print(df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Following and Target Selection using the cleaned DataFrame df
print("Data: After Dataprocessing\n" ,df)
X = df[['Carpet Area', 'Bathroom']] # <- Use multiple features,  # Independent variables
Y = df['Price (in rupees)']  # Dependent variable

print(X, Y)
```

```python
# Step 5: Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

print("X_train : \n",X_train)
print("______________\n")
print("Y_train : \n",Y_train)
print("______________\n")
print("X_test : \n",X_test)
print("______________\n")
print("Actual Y_test : \n",Y_test)

# Linear Regression Model
model = LinearRegression()
model.fit(X_train, Y_train)

print("_____________\n")

# Prediction
pred_Y_test = model.predict(X_test)

# Combine results for test set
combined_df = X_test.copy()
combined_df['Actual_Y_test'] = Y_test
combined_df['Predicted_Y_test'] = pred_Y_test

print("Predicted_Y_test based on X_test:\n", combined_df)

# print("\n Predicted vs Actual:\n", combined_df)
```

```python
# Find the Value of [m in Y = mX + c]
slope = model.coef_[0]
intercept = model.intercept_

# Print Best Fit Line Equation
print(f"\nBest Fit Line Equation: Price (in rupees) = {slope:.3f} * Carpet Area + {intercept:.3f}")  # Format Specification
```

```python
import plotly.express as px

# Make sure your DataFrame column is named "Attendance (%)"
plot_df = pd.DataFrame({
    'Carpet Area': X_test['Carpet Area'],
    'Bathroom': X_test['Bathroom'],
    'Actual_Price (in rupees)': Y_test,
    'Predicted_Price (in rupees)': pred_Y_test
})

# Plotly 3D scatter plot for actual Price (in rupees)
fig = px.scatter_3d(
    plot_df,
    x='Carpet Area',
    y='Bathroom',
    z='Actual_Price (in rupees)',
    color='Predicted_Price (in rupees)',
    title='3D Scatter Plot: Carpet Area vs Bathroom vs Price (in rupees) (Actual)',
    labels={
        'Carpet Area': 'Carpet Area',
        'Bathroom': 'Bathroom',
        'Actual_Price (in rupees)': 'Price (in rupees)'
    },
    opacity=0.7
)
fig.show()
```
